```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("NIFTY50_all.csv")
```

```python
df.isnull().sum()
```

```python
df.Trades.fillna(0, inplace=True)
```

```python
import locale

# Set the locale to Indian English
locale.setlocale(locale.LC_ALL, 'en_IN')


# Format the turnover as Indian currency
df['Turnover'] = df['Turnover'].apply(lambda x: locale.currency(x, grouping=True))

```

```python
df.drop(['Deliverable Volume', '%Deliverble'],axis=1, inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Symbol'] = df['Symbol'].replace('MUNDRAPORT', 'ADANIPORTS', regex=True)
df['Symbol'] = df['Symbol'].replace('BHARTI', 'BHARTIARTL', regex=True)
df['Symbol'] = df['Symbol'].replace('HEROHONDA', 'HEROMOTOCO', regex=True)
df['Symbol'] = df['Symbol'].replace('HINDLEVER', 'HINDUNILVR', regex=True)
df['Symbol'] = df['Symbol'].replace('HINDLEVER', 'HINDUNILVR', regex=True)
df['Symbol'] = df['Symbol'].replace('HINDALC0', 'HINDALCO', regex=True)
df['Symbol'] = df['Symbol'].replace('INFOSYSTCH', 'INFY', regex=True)
df['Symbol'] = df['Symbol'].replace('JSWSTL', 'JSWSTEEL', regex=True)
df['Symbol'] = df['Symbol'].replace('KOTAKMAH', 'KOTAKBANK', regex=True)
df['Symbol'] = df['Symbol'].replace('UNIPHOS', 'UPL', regex=True)
df['Symbol'] = df['Symbol'].replace('SESAGOA', 'VEDL', regex=True)
df['Symbol'] = df['Symbol'].replace('SSLT', 'VEDL', regex=True)
df['Symbol'] = df['Symbol'].replace('ZEETELE', 'ZEEL', regex=True)
```

```python
df1 = df1.append({'Company Name': 'Uti Bank ltd', 'Industry': 'FINANCIAL SERVICES' , 'Symbol':'UTIBANK'}, ignore_index=True)
df1 = df1.append({'Company Name': 'Bajaj Auto Finance ltd', 'Industry': 'FINANCIAL SERVICES' , 'Symbol':'BAJAUTOFIN'}, ignore_index=True)
df1 = df1.append({'Company Name': 'Total Telcom ltd', 'Industry': 'TELECOM' , 'Symbol':'TELCO'}, ignore_index=True)
df1 = df1.append({'Company Name': 'TISCO Financial Group Public Company ltd', 'Industry': 'FINANCIAL SERVICES' , 'Symbol':'TISCO'}, ignore_index=True)
```

```python
df1
```

```python
merged_df = pd.merge(df, df1, on='Symbol')
```

```python
merged_df
```

```python
top_values_by_industry = merged_df.groupby('Industry')['Volume'].mean().nlargest(5)
print("Top 5 Industries by Average Volume:\n",top_values_by_industry)
```

```python
top_values_by_industry1 = merged_df.groupby('Industry')['Trades'].mean().nlargest(5)
print("Top 5 Industries by Aver̥age Trades:\n",top_values_by_industry1)
```

```python
top_values_by_ticker1= merged_df.groupby('Company Name')['Volume'].mean().nlargest(5)
print("Top 5 Companies by Average Volume:\n",top_values_by_ticker1)
```

```python
top_values_by_ticker = merged_df.groupby('Company Name')['Trades'].mean().nlargest(5)
print("Top 5 Companies by Aver̥age Trades:\n",top_values_by_ticker)
```

```python
top_values_by_ticker2 = merged_df.groupby('Company Name')['Trades'].sum().nlargest(5)
print("Top 5 Companies by sum of Trades:\n",top_values_by_ticker2)
```
